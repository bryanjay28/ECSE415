Laplacian of Gaussian


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# load image
image = cv2.imread('./woman.png', cv2.IMREAD_GRAYSCALE)

# take the gaussian blur
woman = cv2.GaussianBlur(image, (3,3), 1)

# apply 5 x 5 laplacian filter to ther intensity channel
I_lap = cv2.Laplacian(woman, cv2.CV_32F, ksize=5)

# get the rows and columns
rows, cols = I_lap.shape[:2]
zeroCrossing = np.zeros(I_lap.shape)

# find the rows and columns 
for x in range(0, rows):
    for y in range(0, cols):
        if(I_lap[x,y] >= 0 and I_lap[x,y] - 1 <= 0):
            zeroCrossing[x,y] = 255
#         if(x + 1 >= rows || x - 1 < 0 || y + 1 >= cols || y - 1 < 0)
#             continue
#         if()


# display images
plt.figure(figsize=(20, 20))
plt.subplot(121), plt.imshow(image, cmap="gray")
plt.title('Orginal Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(I_lap, cmap="gray")
plt.title('Laplacian filtered (5x5) image'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(zeroCrossing, cmap="gray")
plt.title('Zero Edge Crossing image'), plt.xticks([]), plt.yticks([])
plt.show()

```
