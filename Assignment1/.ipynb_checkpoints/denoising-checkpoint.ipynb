### Denoising

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

from skimage import measure

# load images
clean_img_path = './peppers.png'
clean_img = cv2.imread(clean_img_path)

noise_img_path = './gaussian_noise_peppers.png'
noise_img = cv2.imread(noise_img_path)

# filter image using the Gaussian Filter, with std dev sqrt(2)
sqrt_2 = np.sqrt(2)
gaussian_img = cv2.GaussianBlur(noise_img, (7,7), 0, sqrt_2, sqrt_2)

# create a box filter
kernel = np.ones((7,7), np.float32)/(5*5)

# filter image using the box plot filter
box_img = cv2.filter2D(noise_img, -1, kernel)

# convert BGR to RGB
clean_img = cv2.cvtColor(clean_img, cv2.COLOR_BGR2RGB)
noise_img = cv2.cvtColor(noise_img, cv2.COLOR_BGR2RGB)
gaussian_img = cv2.cvtColor(gaussian_img, cv2.COLOR_BGR2RGB)
box_img = cv2.cvtColor(box_img, cv2.COLOR_BGR2RGB)

# Print images
plt.imshow(clean_img), plt.title('Peppers Clean Image')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(noise_img), plt.title('Peppers Original Noisy Image')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(gaussian_img), plt.title('Peppers 7x7 Gaussian')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(box_img), plt.title('Peppers 7x7 Box Filter')
plt.xticks([]), plt.yticks([])
plt.show()

# Calculate the PSNR levels in comparison to the original
gauss_psnr = measure.compare_psnr(clean_img, gaussian_img)
box_psnr = measure.compare_psnr(clean_img, box_img)

print('gauss_psnr:', gauss_psnr)
print('box_psnr:', box_psnr)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

from skimage import measure

# load images
clean_img_path = './peppers.png'
clean_img = cv2.imread(clean_img_path)

snp_img_path = './saltandpepper_noise_peppers.png'
snp_img = cv2.imread(snp_img_path)

# filter image using the Gaussian Filter, with std dev sqrt(2)
sqrt_2 = np.sqrt(2)
gaussian_img = cv2.GaussianBlur(snp_img, (7,7), 0, sqrt_2, sqrt_2)

# filter image using the 7x7 median filter
box_img = cv2.medianBlur(snp_img, 7)

# convert BGR to RGB
clean_img = cv2.cvtColor(clean_img, cv2.COLOR_BGR2RGB)
snp_img = cv2.cvtColor(snp_img, cv2.COLOR_BGR2RGB)
gaussian_img = cv2.cvtColor(gaussian_img, cv2.COLOR_BGR2RGB)
box_img = cv2.cvtColor(box_img, cv2.COLOR_BGR2RGB)

# Print images
plt.imshow(clean_img), plt.title('Peppers Clean Image')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(snp_img), plt.title('Peppers Original Noisy Image')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(gaussian_img), plt.title('Peppers 7x7 Gaussian')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(box_img), plt.title('Peppers 7x7 Median Filter')
plt.xticks([]), plt.yticks([])
plt.show()

# Calculate the PSNR levels in comparison to the original
gauss_psnr = measure.compare_psnr(clean_img, gaussian_img)
box_psnr = measure.compare_psnr(clean_img, box_img)

print('gauss_psnr:', gauss_psnr)
print('box_psnr:', box_psnr)
```
