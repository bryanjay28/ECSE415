### Sharpening

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# load image
rice_path = './rice.png'
image = cv2.imread(rice_path)

# create a box filter 5x5
kernel = np.ones((5,5), np.float32)/(5*5)

# filter using box filter
box_img = cv2.filter2D(image, -1, kernel)

# filter using gaussian filter
sqrt_2 = np.sqrt(2)
gauss_img = cv2.GaussianBlur(image, (5,5), 0, sqrt_2, sqrt_2)

# calculate details for sharpening
box_dets = image - box_img
gauss_dets = image - gauss_img

# calculate sharpened image
box_img_sharp = image + box_dets
gauss_img_sharp = image + gauss_dets

# display images
plt.imshow(image), plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(box_img_sharp), plt.title('Box Filtered Image Sharpened')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(gauss_img_sharp), plt.title('Gaussian Filtered Image Sharpened')
plt.xticks([]), plt.yticks([])
plt.show()
```
