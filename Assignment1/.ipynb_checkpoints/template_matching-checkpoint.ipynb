Template Matching

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

# load all the images
waldo_template1 = cv2.imread('./template1.png')
background_img = cv2.imread('./where_is_waldo.jpg')

# convert the colour for the pictures
waldo_template1 = cv2.cvtColor(waldo_template1, cv2.COLOR_BGR2RGB)
background_img = cv2.cvtColor(background_img, cv2.COLOR_BGR2RGB)

# background copy
background_cpy = background_img.copy()

height, width = waldo_template1.shape[0:2]

# apply template mathcing
res = cv2.matchTemplate(background_img, waldo_template1, cv2.TM_CCOEFF_NORMED)

# set appropriate threshhold
threshold = 0.6

# find all point located above the threshold
points = np.where(res >= threshold)

# loop through all the points 
for point in zip(*points[::-1]):
    cv2.rectangle(background_cpy, point, (point[0]+width, point[1]+height), (0,0,0), 5)
    print(point)
    
# display images
plt.imshow(background_img)
plt.title("Where is waldo board"), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(waldo_template1)
plt.title("Template Image of Waldo"), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(background_cpy)
plt.title("Matched Image"), plt.xticks([]), plt.yticks([])
plt.show()
```
